```python
import math
```

```python
# Perfect sensor
def sense(x):
    return x
```

```python
def simulate(Δt, x, u):
    x += Δt * u
    return x
```

```python
def control(t, y):
    ### WRITE YOUR CONTROL POLICY HERE:
    ux = -sin(t)
    uy = cos(t)
    uz = 1

    return array([ux, uy, uz])
```

```python
tf = 10
Δt = 0.05    # Time step
time = linspace(0.,tf, int(tf / Δt) + 1)  # Time interval


# Initial conditions
x = array([0., 0., 0.])
x_log = [copy(x)]

for t in time:
    y = sense(x)
    u = control(t, y)    
    x = simulate(Δt, x, u)
    x_log.append(copy(x))
    
x_log = array(x_log)
```

```python
grid()
plot( x_log[:,1], x_log[:,2])
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt



fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = x_log[:,0]
y = x_log[:,1]
z = x_log[:,2]



ax.scatter(x, y, z, c='r', marker='o')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
